```python
# plot the solution with enhanced visuals
import matplotlib.pyplot as plt

from data_model import Instance, DropOff

plt.figure(figsize=(12, 10))

# Set up colors for different routes
colors = ["g", "b", "r", "purple", "m", "orange", "brown"]

# plot the depot with a star marker and label
depot_location = instance.locations[instance.depot_location]
plt.scatter(
    depot_location.x,
    depot_location.y,
    s=250,
    c="red",
    marker="*",
    edgecolor="black",
    linewidth=1.5,
    label="Depot",
)


def tour_length(instance: Instance, tour: list[DropOff]) -> float:
    if not tour:
        return 0.0
    length = instance.distance_to_depot(tour[0])
    for i in range(len(tour) - 1):
        length += instance.distance(tour[i].location_id, tour[i + 1].location_id)
    length += instance.distance_to_depot(tour[-1])
    return length


obj = sum(tour_length(instance, route) for route in tour if route)

# plot the customers with better visibility
for i, location in enumerate(instance.locations):
    if i != instance.depot_location:
        plt.scatter(
            location.x,
            location.y,
            s=100,
            c="blue",
            alpha=0.7,
            edgecolor="black",
            linewidth=0.5,
        )
        plt.annotate(str(i), (location.x + 1, location.y + 1), fontsize=8)

# plot each tour with a different color
for i, route in enumerate(tour):
    if not route:
        continue

    color = colors[i % len(colors)]

    # Draw line from depot to first location
    plt.plot(
        [depot_location.x, instance.locations[route[0].location_id].x],
        [depot_location.y, instance.locations[route[0].location_id].y],
        c=color,
        linewidth=1.5,
        alpha=0.8,
    )

    # Draw arrow to show direction
    mid_x = (depot_location.x + instance.locations[route[0].location_id].x) / 2
    mid_y = (depot_location.y + instance.locations[route[0].location_id].y) / 2
    plt.annotate(
        "",
        xy=(
            instance.locations[route[0].location_id].x,
            instance.locations[route[0].location_id].y,
        ),
        xytext=(mid_x, mid_y),
        arrowprops=dict(arrowstyle="->", color=color),
    )

    # Draw lines between locations in the route
    for j in range(len(route) - 1):
        plt.plot(
            [
                instance.locations[route[j].location_id].x,
                instance.locations[route[j + 1].location_id].x,
            ],
            [
                instance.locations[route[j].location_id].y,
                instance.locations[route[j + 1].location_id].y,
            ],
            c=color,
            linewidth=1.5,
            alpha=0.8,
        )

        # Draw arrow to show direction
        mid_x = (
            instance.locations[route[j].location_id].x
            + instance.locations[route[j + 1].location_id].x
        ) / 2
        mid_y = (
            instance.locations[route[j].location_id].y
            + instance.locations[route[j + 1].location_id].y
        ) / 2
        plt.annotate(
            "",
            xy=(
                instance.locations[route[j + 1].location_id].x,
                instance.locations[route[j + 1].location_id].y,
            ),
            xytext=(mid_x, mid_y),
            arrowprops=dict(arrowstyle="->", color=color),
        )

    # Draw line from last location back to depot
    plt.plot(
        [instance.locations[route[-1].location_id].x, depot_location.x],
        [instance.locations[route[-1].location_id].y, depot_location.y],
        c=color,
        linewidth=1.5,
        alpha=0.8,
    )

    # Add a label for the legend
    plt.plot([], [], c=color, label=f"Route {i+1} ({len(route)} stops)")

# Add title and legend
plt.title("Multi-Capacitated Vehicle Routing Solution", fontsize=16)
plt.xlabel("X Coordinate", fontsize=12)
plt.ylabel("Y Coordinate", fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend(loc="upper right", fontsize=10)
plt.axis("equal")
# Add text with summary info with better positioning and formatting
plt.figtext(
    0.02,
    0.02,
    f"Total routes: {sum(1 for r in tour if r)}\nVehicle capacity: {instance.vehicle_capacity}\nTotal distance: {obj:.2f}",
    fontsize=10,
    bbox=dict(facecolor="white", alpha=0.7, edgecolor="gray", boxstyle="round,pad=0.5"),
)

# Adjust layout to prevent overlapping
plt.tight_layout(rect=[0, 0.05, 1, 0.95])
plt.show()
```
